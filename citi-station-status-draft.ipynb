#### Citi Bike Real-Time Data

```python
import requests
import json
import pandas as pd
import pprint
import numpy as np
```

```python
feeds_url = "http://gbfs.citibikenyc.com/gbfs/gbfs.json"
feeds_resp = requests.get(feeds_url)

feeds_content = feeds_resp.text
feeds_data = json.loads(feeds_content)

feeds = pd.DataFrame(feeds_data['data']['en']['feeds'])
```

```python
status_url = "https://gbfs.citibikenyc.com/gbfs/en/station_status.json"
status_resp = requests.get(status_url)

status_content = status_resp.text
status_data = json.loads(status_content)

status = pd.DataFrame(status_data['data']['stations'])
```

```python
import requests
import json

status_url = "https://feeds.citibikenyc.com/stations/stations.json"
status_resp = requests.get(status_url)

status_content = status_resp.text
status_data = json.loads(status_content)

time_suffix = str(status_data["executionTime"])

file_name = "/Users/chuamelia/Downloads/station_status_{0}.json".format(time_suffix)
f = open(file_name,"w+")
f.write(str(status_data))
f.close()
```
